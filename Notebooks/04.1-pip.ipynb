# 04.1 pip

Pip is the official python package management system. It is lightweight and versatile.

Sometimes you may need to work on a remote machine. You have limited access to the remote machine, like only a terminal with no graphical interface. With python and pip, you may be able to configure the remote machine to have all the needed packages you will need.

<div class="alert alert-warning">
    <b>The pip demo is just for you to follow. We will cover basic interactions here of virtual environment creation.<b></br>
    <b>Afterwards, with conda, we will all perform the actions in the demo.<b></br>
<\div>

On the remote machine:

```shell
pip search numpy XXXXXX
```

<div class="alert alert-danger">
    <b>Code Deprecation is something your team must avoid at all costs. Your business model cannot be dependent on code that is no longer maintained!<b></br>
</div>

Even working with software that is no longer supported, you can use the package manager to save your workflows. You know there is another version of python in your system:
    
Make sure what pip version you are using. Enforce "python3" when you are working in a machine with several python versions installed.
    
```shell
python3 -m pip --version
```

Remember pip does not create virtual environments alone. You need to install virtualenv.
    
```shell
python3 -m pip install --user virtualenv
```

Let's create a simple venv calle adpro. We'll make sure it is a python 3 venv. On Mac and Linux
    
```shell
python3 -m venv adpro
```

On Windows:
    
```shell
py -m venv adpro
```

A directory called adpro is created. It contains a barebones python ditributions. Acivate the virtual environment in that directory with:
    
```shell
source ./adpro/bin/
```

As you can see, there is no numpy. We can install it with:    

```shell
pip install numpy
```
Notice the chip set! You might have a different output and/or version.

We can use pip to uninstall numpy as well:
    
```shell
pip uninstall numpy
```

<div class="alert alert-danger">
    <b>Since you can't search for packages, the next commands might also output an error.<b></br>
</div>
    
```shell
pip list --outdated
```
    
```shell    
pip list -o
```    
    
To see what you have installed, run a freeze.
```shell
pip freeze
```

You can output the list into a file.
```shell    
pip freeze > environment.pip
```

Check it:
```shell
cat environment.pip
```

Leave the virtual environment.
```shell
deactivate
```

<div class="alert alert-success">
    <b>You now have a template of a working virtual environment you can share with your colleague. That way you assure everyone has the same packages!<b></br>
</div>
    
If you wanted, you could create a new venv and make sure it had the correct package settings by doing:
    
```shell
pip install -r environment.pip
```
